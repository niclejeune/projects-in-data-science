```python
import pandas as pd
import matplotlib.pyplot as plt
csv_path = '../img_id_and_values.csv'
```

```python
data_frame = pd.read_csv(csv_path)

columns = ['hsv_score', 'contrast', 'homogeneity', 'energy', 'correlation']
diagnostic_values = data_frame['diagnostic'].unique()

colors = {
    'BCC': 'red',
    'MEL': 'red',
    'SCC': 'red',
    'ACK': 'blue',
    'NEV': 'blue',
    'SEK': 'blue',
    'Unknown': 'gray'}

# Loop through each column and create scatter plots
for column in columns:
    plt.figure(figsize=(8, 6))  
    for diagnostic in diagnostic_values:
        subset = data_frame[data_frame['diagnostic'] == diagnostic]
        plt.scatter(subset[column], subset['diagnostic'], label=diagnostic,color=colors.get(diagnostic, 'black') )

    plt.title(f'Scatter Plot: {column} vs diagnostic')
    plt.xlabel(column)
    plt.ylabel('diagnostic')
    plt.legend()
    plt.grid(True)
    plt.show()
```
